````python

a = "a"
b = "b"
c = 1
d = 2

## Penjumlahan (+)
print(f"{c} + {d} = {c + d}") # int + int
print(f"{c} + {a} = {c + a}") # int + str -> error
print(f"{a} + {b} = {a + b}") # str + str

## Pengurangan (-)
print(f"{c} - {d} = {c - d}") # int - int
print(f"{c} - {a} = {c - a}") # int - str -> error
print(f"{a} - {b} = {a - b}") # str - str -> error

## Perkalian (*)
print(f"{c} * {d} = {c * d}") # int * int
print(f"{c} * {a} = {c * a}") # int * str 
print(f"{a} * {b} = {a * b}") # str * str -> error

## Pembagian (/)
print(f"{c} / {d} = {c / d}") # int / int
print(f"{c} / {a} = {c / a}") # int / str -> error
print(f"{a} / {b} = {a / b}") # str / str -> error

````

```
buat program untuk menghitung akar-akar persamaan kuadrat, input a, b, dan c dengan a,b,c bilangan real dan ax^2+bx+c=0, dengan output akar-akar dari persamaan kuadrat tersebut
```



````python
#CODE Here:

print('\nBentuk dari persamaan kuadrat adalah : ax^2+bx+c=0' + "\n")

a = int(input('masukkan nilai a : '))
b = int(input('masukkan nilai b : '))
c = int(input('masukkan nilai c : '))

d = (b**2) - (4*a*c)

print(f"akar-akar dari persamaan kuadrat {a}x^2+{b}x+{c}: ")

if d > 0:
    akar_1 = (-b + (d**0.5)) / (2*a)
    akar_2 = (-b - (d**0.5)) / (2*a)
    print(f'''
    D > 0
    x1 = {akar_1:.0f}
    x2 = {akar_2:.0f}''')

elif d == 0:
    akar = -b / (2*a)
    print(f"D > 0\nx1 = x2 = {akar:.0f}")

elif d < 0:
    akar_1 = (-b + (d**0.5)) / (2*a)
    akar_2 = (-b - (d**0.5)) / (2*a)
    print(f'''
    D < 0
    x1 = {akar_1:.0f}
    ''')

print("Catatan: yang saya ambil hanya floornya (angka dibelakang koma tidak ditampilkan)")
````
